## Dictionaries

### What is a Dictionary

A dictionary stores data in key value pairs.
Keys must be unique.

### Creating a Dictionary

```python
student = {
    "Name": "Joy",
    "Age": 16,
    "Gender": "Female"
}

print(student)
```

Accessing Values

```python
print(student["Name"])
print(student["Gender"])
```

Changing Values

```python
student["Age"] = 17

print(student)
print(student["Age"])
```

### Adding new key value pair

```python
student["Score"] = 98

print(student)
```

### Removing Items

```python
student.pop("Name")

print(student)
```
