```python

# IMPORT LIBRARIES
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd
```

```python
# CREATE DATASET
data = {
    "hours_studied": [1, 2, 3, 4, 5, 6, 7, 8],
    "exam_score": [50, 55, 65, 70, 75, 85, 90, 95]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)
```

```python

# DEFINE FEATURES (X) AND TARGET (Y)
x = df[["hours_studied"]]
y = df["exam_score"]
# SPLIT DATA
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python


# ==============================
# TRAIN MODEL
# ==============================

David = LinearRegression()
David.fit(x_train, y_train)


# ==============================
# MAKE PREDICTION
# ==============================

prediction = David.predict([[9]])
print("\nPredicted score for 9 hours study:", prediction)


# ==============================
# CHECK MODEL PERFORMANCE
# ==============================

y_pred = David.predict(x_test)

print("\nTest Data:")
print(x_test)

print("\nPredictions:")
print(y_pred)

print("\nActual Scores:")
print(y_test)

error = mean_squared_error(y_test, y_pred)
print("\nMean Squared Error:", error)
```
