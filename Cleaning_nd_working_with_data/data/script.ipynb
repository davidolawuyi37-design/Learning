```python
import pandas as pd 
```

```python
df = pd.read_csv("dave.csv")
print(df)
```

```python
df.columns = df.columns.str.strip() # Remove white space from the columns
df["Name"] = df["Name"].str.strip().str.capitalize() # Remove white space from the Name columns
df["Course"] = df["Course"].str.strip().str.lower() # Remove white space from the Course  columns
print(df)
print(df)
```

```python
# Standardize Course names
df["Course"] = df["Course"].replace({
    "ai": "AI", 
    "cybersecurity": "Cybersecurity",
    "data science": "Data science"
})
print(df)
```

```python
# Handling missing values (Dropna & Fillna)
# df = df.dropna()

df["Age"] = df["Age"].fillna(df["Age"].median())
df["Score"] = df["Score"].fillna(df["Score"].mean())
df["Course"] = df["Course"].fillna("Unknown")
df["Name"] = df["Name"].fillna("Unknown")
print(df)
```

```python
# Fix data types
df["Age"] = df["Age"].astype(int)
df["Score"] = df["Score"].astype(int)
print(df)
```

```python
# Remove duplicates
df = df.drop_duplicates()
print(df)
```

```python
# Save Clean Data
import os
os.makedirs("Cleaning_nd_working_with_data/data", exist_ok=True)
df.to_csv("Cleaning_nd_working_with_data/data/cleaned.csv", index=False)
print("Data saved successfully.")
```
