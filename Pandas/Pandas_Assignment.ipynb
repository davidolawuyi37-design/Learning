```python
import pandas as pd
import numpy as np
```

#### Create the Data Frame

```python
data = {
    "Student": ["David", "Esther", "Michael", "Grace", "Daniel"],
    "Age": [18, 19, 17, 20, 18],
    "Math": [78, 85, 69, 90, 74],
    "English": [82, 88, 75, 91, 79],
    "Science": [80, 92, 70, 89, 76]
}
df = pd.DataFrame(data)
```

##### Part 1: Exploring the Data

```python
print("First 3 rows:")
print(df.head(3), "\n")
```

```python
print("Last 2 rows:")
print(df.head(2), "\n")
```

```python
print("Shape of DataFrame:")
print(df.shape, "\n")
```

```python
print("Column names:")
print(df.columns.tolist(), "\n")
```

##### Part 2: Working with columns

```python
print("Student and Math Columns:")
print(df[["Student", "Math"]], "\n")
```

```python
print("Average score for each subject:")
print(df[["Math", "English", "Science"]].mean(), "\n")
```

##### Add Total Column

```python
df["Total"] = df["Math"] +  df["English"] + df["Science"]
print("DataFrame with Total Column:")
print(df, "\n")
```

#### Part 3: Data Filtering

```python
print("Students who scored more than 80 in Math:")
print(df[df["Math"] > 80], "\n")

print("Students with total score greater than 240:")
print(df[df["Total"] > 240], "\n")
```

#### Part 4: Data Modification

##### Increase all English scores by 5

```python
df["English"] += 5
```

#### Change David's Science score to 85

```python
df.loc[df["Student"] == "David", "Science"] = 85
```

#### Add Passed column (average >= 75)


```python
df["Average"] = df[["Math", "English", "Science"]].mean(axis=1)
df["Passed"] = df["Average"] >= 75

print("Modified DataFrame:")
print(df, "\n")
```

#### Part 5: Simple Analysis

```python
print("Highest total score:")
print(df["Total"].max(), "\n")

print("Student with lowest Math score:")
print(df.loc[df["Math"].idxmin(), "Student"], "\n")

print("Class average for Science:")
print(df["Science"].mean())
```
