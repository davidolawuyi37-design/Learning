### What is Pandas?


Pandas is a python library used for data manipulation and data analysis.
It is built on top of Numpy and makes working with structured data easy.

```python
import pandas as pd
```

### Series Vs DataFrame
Series:
A one dimensional labeled array.

```python
a = ["a", "b", "c", "d", "e"]
pd.Series(a)
```

DataFrame: A two dimensional table of rows and columns

```python
data = {
    "Name": ["David", "Joy", "Ben", "Jude"],
    "Age": [18, 17, 15, 10],
    "Score": [85, 48, 43, 90]
}
df = pd.DataFrame(data)
print(df)
```

### Veiwing Data


```python
df.head() # Prints the first five rows
df.tail() # Prints the last five rows
df.shape # Prints the shape of the dataFrame
df.columns # Prints the columns head
df.info() # Prints the overall info of the table.
df.describe() # Statistical report of the table
```

### Selecting Data

```python
# Selecting Columns
df["Name"]
df[["Name", "Age", "Score"]]

print(df)

# Selecting Rows
print(df.loc[0, "Age"]) # loc is for custom index
print(df.iloc[0, 1]) # iloc is for original index
```

```python
print(df.iloc[1, 0])
```

```python
print(df.iloc[2, 2])
```

```python
print(df.loc[2, "Age"])
```

### Filtering 

```python
print(df)
```

```python
print(df["Age"] > 16)
```

```python
under_16 = df[df["Age"] < 16]
print(under_16)
```

```python
over_16 = df[df["Age"] > 16]
print(over_16)
```

```python
above_50 = df[df["Score"] > 50]
print(above_50)
```

```python
below_15_above_50 = df[(df["Age"]< 15) & (df["Score"]> 50)]
print(below_15_above_50)
```

```python
below_18_below_50 = df[(df["Age"]< 18) & (df["Score"]< 50)]
print(below_18_below_50)
```
