```python
import pandas as pd
```

```python
df = pd.read_csv("Data/raw/raw.csv")
print(df)
```

```python
df.columns = df.columns.str.strip() # Remove white space from the columns
df["Name"] = df["Name"].str.strip().str.capitalize() # Remove white space from the Name column
df["Course"] = df["Course"].str.strip().str.lower() # Remove whitespace from the Course column

# Standardize Course Names
df["Course"] = df["Course"].replace({

    "ai": "AI",
    "data science": "Data Science",
    "cybersecurity": "Cybersecurity"
})

# Handling missing Values (Droppna/Fillna)
# df = df.dropna()

df["Age"] = df["Age"].fillna(df["Age"].median())
df["Score"] = df["Score"].fillna(df["Score"].mean())
df["Course"] = df["Course"].fillna("Unknown")
df["Name"] = df["Name"].fillna("Unknown")

# Fix Data Types
df["Age"] = df["Age"].astype(int)
df["Score"] = df["Score"].astype(int)
# Do that of score

# Remove Duplicates
df = df.drop_duplicates()


df

# Save Clean Data
df.to_csv("Data/processed/cleaned.csv", index=False)
print
```
