What is a List?

A list is a collection of items stored in one variable.

Creating a List

```python
fruits = ["apple", "bannana", "oranges"]
numbers = [10, 20, 30]
mixed = ["David", 10, True]

print(fruits)
print(numbers)
print(mixed)
```

Accessing Items

```python
print(fruits[0])
print(numbers[2])
print(f"{mixed[0]} is a smart student and this is {mixed[2]}")
```

Changing Items

```python
mixed[0] = "Divine"
print(mixed[0])
print(mixed)
print(f"{mixed[0]} is a smart student and this is {mixed[2]}")
```

Adding Items

```python
mixed.append("Sam")
print(mixed)

mixed.insert(1, "David")
print(mixed)
```

Removing Items

```python
mixed.remove("Divine")
print(mixed)

mixed.pop()
print(mixed)
```
