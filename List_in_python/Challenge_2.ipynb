```python
Numbers = [4, 2, 4, 5, 2, 8, 8, 8, 1]

Unique_nums = []
for Number in Numbers:
    if Number not in Unique_nums:
        Unique_nums.append (Number)

print(Unique_nums)
```

```python
Letters = ["a", "b", "c"]
Numbers = [1, 2. 3]
Combined = []
for l, n in zip(Letters, Numbers):
    Combined.append(l)
    Combined.append(n)
print(Combined)
```
