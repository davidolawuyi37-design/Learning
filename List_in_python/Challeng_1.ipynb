## List Challenge 1

1. Create a list of four Nigerian cities and print the list.

2. Add one new city to the end of the list and print the updated list.

3. Change the second item in the list to a different city and print the result.

4. Remove one city from the list and print what is left.

5. Print the length of the list.

```python
Nigerian_Cities = ["Lagos","Abuja","Ilorin","Port-Hacourt", "Kano"]
print(Nigerian_Cities)
```

```python
Nigerian_Cities.append("Sokoto")
print(Nigerian_Cities)
```

```python
Nigerian_Cities = ["Lagos","Abuja","Ilorin","Port-Hacourt", "Kano"]
Nigerian_Cities[2] ="Benue"
print(Nigerian_Cities)
```

```python
Nigerian_Cities.remove("Lagos")
print(Nigerian_Cities)
```

```python
print(len(Nigerian_Cities))
```
