### What is Numpy?
Numpy stands for Numeric PYthon.
It is a core library for numeric computation in python and the foundation of data science, machine learning, and AI workflows.

### Why is Numpy importatant?

- Much faster than Python lists
- Uses less memory
- Supports vectorized operations
- Designed for scientific computation

```python
import numpy as np

Numbers = [1, 2, 3, 4, 5]
Array = np.array([1, 2, 3, 4, 5])

print(Numbers)
print(Array)
converted = np.array(Numbers)
print(converted)
```
