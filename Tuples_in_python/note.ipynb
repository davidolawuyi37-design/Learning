## What is a Tuple?
A turple is a collection of items just like a list, but the key difference is:
### Tuples cannot be changed (modified).
This means:
1. You cannot add new items
2. You cannot remove items
3. You cannot change any item

## What is a tuple now used for?

Tuples are useful when you want data that should not be modified.

## Creating a Tuple

```python
colors = ("red", "blue", "green")
colors
```

### Accessing Items

```python
print(colors[1])
print(colors[0])
```

### Tuple Length

```python
print(len(colors))
```

### Mixed Data

```python
person = ("David", 16, "Nigeria")
print(person)
```
