```python
Fruits = ("Pineapple", "Grape", "Avocado", "Sour Sap", "Watermelon", "Cashew")
print(Fruits[0])
print(Fruits[5])
```

```python
Numbers = (5, 23, 33, 45, 10, 7)
print(min(Numbers))
print(max(Numbers))   
```
